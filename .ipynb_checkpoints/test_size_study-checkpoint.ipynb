## Data wrangling

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from tqdm.auto import tqdm

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data',header=None)
columns = ["age","sex","cp","trestbps","chol","fbs","restecg","thalach","exang","oldpeak","slope","ca","thal","num"]   
df.columns = columns    
df = df.drop(df.loc[(df == '?').any(axis=1)].index)
df['ca'] = df['ca'].astype(float)
df['thal'] = df['thal'].astype(float)
y = df['num'].copy()
X = df.drop('num',axis=1).copy()
X_encoded = pd.get_dummies(X,columns = ["cp","restecg","slope","ca","thal"]) # one-hot-encoding
X_encoded["exang"] = X_encoded["exang"].astype('uint8')
X_encoded["fbs"] = X_encoded["fbs"].astype('uint8')
X_encoded["sex"] = X_encoded["sex"].astype('uint8')
y_mask = y > 0
y[y_mask] = 1
```

## Varying the test size - repeated hold-out

```python
pipe_lr = make_pipeline(
    StandardScaler(),
    PCA(),
    LogisticRegression(max_iter = 1000,fit_intercept=True,solver='liblinear')
)

X_val = X_encoded.copy().values
y_val = y.copy().values

means_2000 = []
stds_2000 = []
test_sizes_2000 = []
for i in range(50):
  test_size = (50-i)/100
  print(test_size)
  acc = []
  for i in tqdm(range(2000)):
    x_train, x_test, y_train, y_test = train_test_split(X_encoded, y, stratify=y,test_size = test_size, random_state = i)
    pipe_lr.fit(x_train,y_train)
    acc.append(pipe_lr.score(x_test,y_test))
  means_2000.append(np.mean(acc))
  stds_2000.append(np.std(acc,ddof=1))
  test_sizes_2000.append(test_size)
means_2000 = pd.Series(means_2000)
stds_2000 = pd.Series(stds_2000)
test_sizes_2000 = pd.Series(test_sizes_2000)
```

## Functions

```python
def guaranteed_accuracy(nr_models,means_2000,stds_2000,test_sizes_2000):
  means_2000 = means_2000[:-2].rolling(3).mean()[2:] # test size < 4% is too biased, rolling average to combat variance
  stds_2000 = stds_2000[1:-3] # matching indices with means_2000
  test_sizes_2000 = test_sizes_2000[1:-3] # matching indices with means_2000
  loss = pd.Series([i - 2*j*np.sqrt(1/nr_models+(nr_models-1)/nr_models*k/(1-k)) for i,j,k in zip(means_2000,stds_2000,test_sizes_2000)],index=test_sizes_2000.values)
  return max(loss),loss.idxmax() # returns highest guaranteed accuracy and corresponding test_size

def supported_accuracy(nr_models,means_2000,stds_2000,test_sizes_2000):
  means_2000 = means_2000[:-2].rolling(3).mean()[2:] # test size < 4% is too biased, rolling average to combat variance
  stds_2000 = stds_2000[1:-3] # matching indices with means_2000
  test_sizes_2000 = test_sizes_2000[1:-3] # matching indices with means_2000
  loss = pd.Series([i - j*np.sqrt(1/nr_models+(nr_models-1)/nr_models*k/(1-k)) for i,j,k in zip(means_2000,stds_2000,test_sizes_2000)],index=test_sizes_2000.values)
  return max(loss),loss.idxmax() # returns highest supported accuracy and corresponding test_size
```

## First Plot

```python
# first subplot

nr_models_ = np.linspace(1,100,100)
opt_test_ = [guaranteed_accuracy(i,means_2000,stds_2000,test_sizes_2000)[1] for i in nr_models_]
plt.plot(nr_models_,opt_test_, label = 'maximizing guaranteed accuracy')
opt_test_ = [supported_accuracy(i,means_2000,stds_2000,test_sizes_2000)[1] for i in nr_models_]
plt.plot(nr_models_,opt_test_, label = 'maximizing supported accuracy')
plt.plot(nr_models_[1:],[1/i for i in nr_models_][1:], label = 'k-fold cv comparison',color='brown')
plt.ylim(bottom = 0.0)
plt.xlabel('# of models')
plt.ylabel('optimal test size')
plt.legend()
plt.grid()
plt.tight_layout()
plt.xscale('log')
plt.show()

# second subplot

nr_models_ = np.linspace(1,100,100)

acc_gain_0 = guaranteed_accuracy(1,means_2000,stds_2000,test_sizes_2000)[0]
acc_gain_ = [guaranteed_accuracy(i,means_2000,stds_2000,test_sizes_2000)[0]- acc_gain_0 for i in nr_models_]
plt.plot(nr_models_,acc_gain_, label = 'guaranteed accuracy')
acc_gain_0 = supported_accuracy(1,means_2000,stds_2000,test_sizes_2000)[0]
acc_gain_ = [supported_accuracy(i,means_2000,stds_2000,test_sizes_2000)[0]- acc_gain_0 for i in nr_models_]
plt.plot(nr_models_,acc_gain_, label = 'supported accuracy')

test_sizes_ = [guaranteed_accuracy(i,means_2000,stds_2000,test_sizes_2000)[1] for i in nr_models_]
means_ = np.array([float(means_2000.loc[test_sizes_2000.loc[test_sizes_2000 == i].index].values) for i in test_sizes_])
plt.plot(nr_models_,means_-means_[0],label='mean accuracy - maximizing guaranteed accuracy')
test_sizes_ = [supported_accuracy(i,means_2000,stds_2000,test_sizes_2000)[1] for i in nr_models_]
means_ = np.array([float(means_2000.loc[test_sizes_2000.loc[test_sizes_2000 == i].index].values) for i in test_sizes_])
plt.plot(nr_models_,means_-means_[0],label='mean accuracy - maximizing supported accuracy')

plt.xlabel('# of models')
plt.ylabel('gain in accuracy')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

## Varying the test size - k-fold cv

```python
from sklearn.model_selection import StratifiedKFold

pipe_lr = make_pipeline(
    StandardScaler(),
    PCA(),
    LogisticRegression(max_iter = 1000,fit_intercept=True,solver='liblinear')
)

X_val = X_encoded.copy().values
y_val = y.copy().values

folds_cv=[]
means_cv = []
stds_cv = []
old_folds = 0
for pts_per_split in tqdm(range(2,int(len(y_val)/2))): # we need at least two data points for stratified cv
  folds = int(np.round(len(y_val)/pts_per_split))
  if folds != old_folds: # we don't need to run the same number of folds twice
    print(folds)
    means = []
    stds = []
    for i in range(int(10*len(y)/folds)): # running '10x the number of test data points' random seeds
      kf = StratifiedKFold(n_splits=folds, shuffle = True, random_state=i)
      acc = []
      for train_index, test_index in kf.split(X_val,y_val):
        X_train, X_test = X_val[train_index], X_val[test_index]
        y_train, y_test = y_val[train_index], y_val[test_index]
        pipe_lr.fit(X_train,y_train)
        acc.append(pipe_lr.score(X_test,y_test))
      means.append(np.mean(acc))
      stds.append(np.std(acc,ddof=1))
    means_cv.append(np.mean(means))
    stds_cv.append(np.mean(stds))
    folds_cv.append(folds)
    old_folds = folds
```

```python
from sklearn.model_selection import KFold

pipe_lr = make_pipeline(
    StandardScaler(),
    PCA(),
    LogisticRegression(max_iter = 1000,fit_intercept=True,solver='liblinear')
)

X_val = X_encoded.copy().values
y_val = y.copy().values

folds_cv_no_strat=[]
means_cv_no_strat = []
stds_cv_no_strat = []
old_folds = 0
for pts_per_split in tqdm(range(1,int(len(y_val)/2))):
  folds = int(np.round(len(y_val)/pts_per_split))
  if folds != old_folds: # we don't need to run the same number of folds twice
    print(folds)
    means = []
    stds = []
    for i in range(int(10*len(y)/folds)): # running '10x the number of test data points' random seeds
      kf = KFold(n_splits=folds, shuffle = True, random_state=i)
      acc = []
      for train_index, test_index in kf.split(X_val,y_val):
        X_train, X_test = X_val[train_index], X_val[test_index]
        y_train, y_test = y_val[train_index], y_val[test_index]
        pipe_lr.fit(X_train,y_train)
        acc.append(pipe_lr.score(X_test,y_test))
      means.append(np.mean(acc))
      stds.append(np.std(acc,ddof=1))
    means_cv_no_strat.append(np.mean(means))
    stds_cv_no_strat.append(np.mean(stds))
    folds_cv_no_strat.append(folds)
    old_folds = folds
```

## Second Plot

```python
plt.plot(folds_cv,means_cv,label = 'stratified k-fold cross validation')
plt.plot(folds_cv_no_strat,means_cv_no_strat,label = 'non-stratified k-fold cross validation')
nr_models_ = np.linspace(1,300,300)
test_sizes_ = [guaranteed_accuracy(i,means_2000,stds_2000,test_sizes_2000)[1] for i in nr_models_]
means_ = np.array([float(means_2000.loc[test_sizes_2000.loc[test_sizes_2000 == i].index].values) for i in test_sizes_])
plt.plot(nr_models_,means_,label='repeated hold-out - maximizing guaranteed accuracy')
nr_models_ = np.linspace(1,300,300)
test_sizes_ = [supported_accuracy(i,means_2000,stds_2000,test_sizes_2000)[1] for i in nr_models_]
means_ = np.array([float(means_2000.loc[test_sizes_2000.loc[test_sizes_2000 == i].index].values) for i in test_sizes_])
plt.plot(nr_models_,means_,label='repeated hold-out - maximizing supported accuracy')
plt.xlabel('# of models/folds')
plt.ylabel('mean accuracy')
plt.xscale('log')
plt.legend()
```
